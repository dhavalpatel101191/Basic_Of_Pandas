*---------------Handling Missing Data with fillna,dropna,interpolate function --------*

```python
Miss_Data = pd.read_csv('../Sample Data/Miss_Data.csv',parse_dates=['day'])     # parse_dates set date column type as timestamp
```

```python
Miss_Data.head()
```

```python
Miss_Data.set_index('day',inplace=True)
```

```python
# Fill NaN with some value
New_df = Miss_Data.fillna(0)            # Replace NaN with 0
```

```python
New_df
```

```python
New_df = Miss_Data.fillna(
    {
        'temperature':0, 
        'windspeed':0, 
        'event':'No Event'
    }
)
```

```python
New_df
```

```python
# Fill NaN with ffill and bfill  caryy forward value from previous or next value
New_df = Miss_Data.fillna(method='ffill')
# New_df = Miss_Data.fillna(method='bfill)
New_df
```

```python
# Argument with ffill
New_df = Miss_Data.fillna(method='ffill',limit=1)
```

```python
New_df
```

```python
# Interpolate Method 
New_df = Miss_Data.interpolate()
```

```python
New_df
```

```python
New_df = Miss_Data.interpolate(method='time')
```

```python
New_df
```

```python
# Drop Missing Values
 # Drop Row which has NaN values
New_df = Miss_Data.dropna()    
New_df
```

```python
# Drp row if it has all NaN
New_df = Miss_Data.dropna(how='all')
New_df
```

```python
# Drop column based on non NaN values 
New_df = Miss_Data.dropna(thresh=1)     # row has at least one value than keep it.
```

```python
New_df
```

*----------------------Concat Dataframe---------------------*

```python
c_weather = pd.DataFrame({
    'City':['Toronto','Ottawa','North York'],
    'Temperature':[25,28,30], 
    'Humidity':[45,50,55]
})

c_weather
```

```python
a_weather = pd.DataFrame({
    'City':['Richmond','Florida','LA'],
    'Temperature':[20,28,30], 
    'Humidity':[40,55,60]
})
a_weather
```

```python
# Concat data frames
df = pd.concat([c_weather,a_weather])
```

```python
df
```

```python
# get continue index
df = pd.concat([c_weather,a_weather],ignore_index=True)
```

```python
df
```

```python
# Get data from particular dataframe
df = pd.concat([c_weather,a_weather],keys=['Canada','US'])
```

```python
df
```

```python
df.loc['Canada']
```

```python
c_weather_temp = pd.DataFrame({
    'City':['Toronto','Ottawa','North York'],
    'Temperature':[25,28,30]
},index=[0,1,2])

c_weather_hum = pd.DataFrame({
    'City':['Ottawa','North York'],
    'Humidity':[50,55]
},index=[1,2])
```

```python
# Append data on side

df1 = pd.concat([c_weather_temp,c_weather_hum],axis=1)
```

```python
df1
```

```python
# Append Series on DataFrame
s = pd.Series(['Humid','Dry','Rain'],name='Event')

df2 = pd.concat([c_weather_temp,s],axis=1)
```

```python
df2
```

*----------------------Merge DataFames----------------*

```python
c_temp = pd.DataFrame({
    'City':['Toronto','Ottawa','North York'],
    'Temperature':[25,28,30]
})

c_hum = pd.DataFrame({
    'City':['Toronto','Ottawa','North York'],
    'Humidity':[50,55,45]
})
```

```python
df3 = pd.merge(c_temp,c_hum,on='City')
```

```python
df3
```
