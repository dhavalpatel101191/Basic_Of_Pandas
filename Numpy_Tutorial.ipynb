```python
# Joining Numpy Arrays
# vstack(), hstack(), column_stack()

n7 = np.array([range(1,11)])
n8 = np.array([range(11,21)])
n9 = np.array([range(21,31)])

n10 = np.vstack((n7,n8,n9))
n11 = np.hstack((n7,n8))
n12 = np.column_stack((n7,n8,n9))

```

```python
# Intersection and Difference (Find common Values and Different Values)
# intersect1d()   setdiff1d(first, second)---> Give unique value of First Array

p1 = np.array([range(10,70,10)])
p2 = np.array([range(50,100,10)])   

n13 = np.intersect1d(p1,p2)   # Give common Values
n14 = np.setdiff1d(p1,p2)     # Give Unique value in p1
n15 = np.setdiff1d(p2,p1)     # Give unique value in p2
```

```python
# Mathematics Function in Numpy

np.sum(n7)
np.min(n7)
np.max(n7)
np.median(n7)
np.mean(n7)
np.sum([n7,n8])
np.sum([n7,n8],axis = 0)
np.sum([n7,n8],axis = 1)
np.std(n7)
```
